### time usec may be weights or time spent on a title

```python
df.sort_values(by='time_usec',ascending=False,inplace=True)
df
```

### this is embarassing

```python
df_by_time = df[['title','time_usec']].groupby(['title']).sum().sort_values(by='time_usec',ascending=False)[:50]
```

```python
df_by_time
```

```python
df_by_time.reset_index(inplace=True)
```

### first get common topics with out common words and punctuations.

```python
df_by_time.title = df_by_time.title.str.replace('[^a-zA-Z]', ' ')
df_by_time.title = df_by_time.title.apply(lambda x: " ".join([item for item in x.split() if item.lower() not in stopwords.words('english')]))
```

```python
df_by_time.title
```

```python
common_topics = pd.Series(' '.join(df_by_time.title.str.replace('[^a-zA-Z]', ' ')).lower().split()).value_counts()[:50]
```

```python
common_topics.index
```

### split titles into entries where common topics appear in title

```python
df_by_time.title = df_by_time.title.str.lower().replace('[^a-zA-Z]', ' ')
df_by_time.title = df_by_time.title.apply(lambda x: " ".join([item for item in x.split() if item.lower() not in stopwords.words('english')]))
df_by_time.title = df_by_time.title.str.split().apply(lambda element:element+['' for _ in range(len(df_by_time.title.str.split().max()))])
```

```python
df_by_time
```

```python
dft_agg = df_by_time[pd.DataFrame(df_by_time.title.tolist()).isin(common_topics.index[:10]).any(1)]
dft_agg
```

###  to build dictionary and tabulate scores

```python
dic = {topic:0 for topic in common_topics.index.tolist()}
for title in dft_agg.title:
    for topic in set(title).intersection(set(common_topics.index.tolist())):
        dic[topic] += dft_agg.loc[dft_agg.title.str.join(sep=" ") == " ".join(title),'time_usec'].values
```

```python
dic
```

### time spent on each most common topic tabulated and sorted

```python
scores = pd.DataFrame(dic).T.sort_values(by=0,ascending=False)
scores
```

```python
scores['scores_std'] = (scores.iloc[:,0] - scores.iloc[:,0].mean())/scores.iloc[:,0].std()
scores
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
scores[0] = scaler.fit_transform(scores[0].to_numpy().reshape(-1,1))
scores.columns = ['min_max_std','scores_std']
```
