```python
import pandas as pd
import numpy as np
import json

with open('/home/pysolver33/Desktop/Takeout/Chrome/BrowserHistory.json') as json_file:
    data = json.load(json_file)
```

```python
df = pd.DataFrame(data['Browser History'])
df.head()
```

### First find repeating unique values


```python
df.title.unique()
```

### Then find most occuring words in titles with out punctuation

```python
titles = df.title.str.replace('[^a-zA-Z]', ' ')
pd.Series(' '.join(titles).lower().split()).value_counts()[:50]
```

### readjust filter with with out common words

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
titles = titles.apply(lambda x: " ".join([item for item in x.split() if item not in stopwords.words('english')]))
```

```python
titles
```

```python
pd.Series(' '.join(titles).lower().split()).value_counts()[:50]
```

### time usec may be weights or time spent on a title

```python
df.sort_values(by='time_usec',ascending=False,inplace=True)
df
```

### standardize time spend on entry

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
df.time_usec = scaler.fit_transform(df.time_usec.to_numpy().reshape(-1,1))
```

```python
df.time_usec
```

```python
df
```

### this is embarassing

```python
df_by_time = df[['title','time_usec']].groupby(['title']).sum().sort_values(by='time_usec',ascending=False)[:50]
```

```python
df_by_time.reset_index(inplace=True)
```

```python
df.loc[df.title == '[Alucardou] Lustful body I & II [Booth]',['title','time_usec']]
```

### finally first taking the 50 entries with time most spend, filter out common words and punctuation from titles and finally do a word count to get most viewed "topics"

```python
df_by_time.title.str.replace('[^a-zA-Z]', ' ')
df_by_time.title = df_by_time.title.apply(lambda x: " ".join([item for item in x.split() if item not in stopwords.words('english')]))
```

```python
pd.Series(' '.join(df_by_time.title.str.replace('[^a-zA-Z]', ' ')).lower().split()).value_counts()[:50]
```
